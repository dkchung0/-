```python
data3 = data66.iloc[:,[0,1]]
```

```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalerA = StandardScaler().fit(data3)
X_scaled = scalerA.transform(data3)
data3 = X_scaled
```

```python
kmeans = KMeans(n_clusters=100)
kmeans.fit(data3)
new_dy = kmeans.predict(data3)

plt.rcParams['font.size'] = 10
fig = plt.figure(figsize=(50,20))

plt.scatter(data3[:,0],data3[:,1], c=new_dy, cmap=plt.cm.spring)
plt.xlabel("經度",Fontproperties="Microsoft JhengHei",fontsize=30)
plt.ylabel("緯度",Fontproperties="Microsoft JhengHei",fontsize=30)

plt.show()
```
