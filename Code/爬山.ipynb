```python
def new_data (data) :
    
    
    #資料前處理
    
    data.經度 = data.經度.round(2)
    data.緯度 = data.緯度.round(2)
    
    e = (data.hms / 10000).apply(np.floor) * 3600
    r = ((data.hms / 10000 - (data.hms / 10000).apply(np.floor) ) * 100) .apply(np.floor) * 60
    t  = (data.hms / 100 - (data.hms / 100) .apply(np.floor)) * 100

    v = e + r + t
    data.hms = v
    
    
    
    # 跑x月的每日
    
    OK =  pd.DataFrame([])
    for i in data.date_upd.unique() :
        b = data.groupby("date_upd").get_group(i)
        a = data.groupby("date_upd").get_group(i)[["經度","緯度"]].groupby(["經度","緯度"])
        
        
        # 找出經緯度去分人
        ANS = np.zeros((len(a),2))
        for j in range(len(a)) :
            ANS[j] = (a.size().index[j])
        
        HP =  pd.DataFrame([])
        
        for k in range(len(ANS)-1) :
            g = b[(b.經度==ANS[k,0])&(b.緯度==ANS[k,1])]
            g = g.iloc[:,[0,1,2,3,7,8,10]]
            
            h = np.zeros(8)
            
            h[0] = np.mean(g.經度)
            h[1] = np.mean(g.緯度)
            h[2] = np.mean(g.海拔高度)
            h[3] = np.mean(g.順時速度)
            h[4] = np.mean(g.date)
        
            h[5] = np.max(g.hms) - np.min(g.hms) 
            if float(h[5]) > 57600 :
                h[5] = 86400 -  float(h[5])
                
            h[6] = (np.max(g.hms_upd) - np.min(g.hms_upd)) < 10000
            h[7] = (np.max(g.date) - np.min(g.date)) == 0
            
            h = pd.DataFrame(h).T
            
            HP = HP.append(h)
            
        
        OK = OK.append(HP)
    
    
    return(OK)

    
```

```python
def new_data (data) :
    
    
    #資料前處理
    
    data.經度 = data.經度.round(2)
    data.緯度 = data.緯度.round(2)
    
    e = (data.hms / 10000).apply(np.floor) * 3600
    r = ((data.hms / 10000 - (data.hms / 10000).apply(np.floor) ) * 100) .apply(np.floor) * 60
    t  = (data.hms / 100 - (data.hms / 100) .apply(np.floor)) * 100

    v = e + r + t
    data.hms = v
    
    
    
    # 跑x月的每日
    
    OK =  pd.DataFrame([])
    for i in data.date_upd.unique() :
        b = data.groupby("date_upd").get_group(i)
        a = data.groupby("date_upd").get_group(i)[["經度","緯度"]].groupby(["經度","緯度"])
        
        
        # 找出經緯度去分人
        ANS = np.zeros((len(a),2))
        for j in range(len(a)) :
            ANS[j] = (a.size().index[j])
        
        HP =  pd.DataFrame([])
        
        for k in range(len(ANS)-1) :
            g = b[(b.經度==ANS[k,0])&(b.緯度==ANS[k,1])]
            g = g.iloc[:,[0,1,2,3,7,8,10]]
            
            h = np.zeros(8)
            
            h[0] = np.mean(g.經度)
            h[1] = np.mean(g.緯度)
            h[2] = np.mean(g.海拔高度)
            h[3] = np.mean(g.順時速度)
            h[4] = np.mean(g.date)
        
            h[5] = np.max(g.hms) - np.min(g.hms) 
            if float(h[5]) > 57600 :
                h[5] = 86400 -  float(h[5])
                
            h[6] = (np.max(g.hms_upd) - np.min(g.hms_upd)) < 10000
            h[7] = (np.max(g.date) - np.min(g.date)) <= 1
            
            h = pd.DataFrame(h).T
            
            HP = HP.append(h)
            
        
        OK = OK.append(HP)
    
    
    return(OK)
```

```python
M = new_data (data)
```

```python
M.reset_index(inplace=True)
M = M.iloc[:,1:]
M
```
